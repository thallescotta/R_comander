**c) Verifique as suposições para a realização do teste t para amostras independentes.**

Antes de realizar o teste t para amostras independentes, é importante verificar algumas suposições. Aqui estão as principais suposições para o teste t em amostras independentes:

*   **Normalidade**: As amostras devem seguir uma distribuição aproximadamente normal. Você pode verificar isso visualmente usando histogramas ou graficamente com um gráfico Q-Q (quantil-quantil). Além disso, para amostras grandes, o teste t é robusto o suficiente para lidar com desvios moderados da normalidade.
*   **Homogeneidade de Variância (Homocedasticidade)**: As variâncias das duas amostras devem ser aproximadamente iguais. Você pode verificar isso visualmente com gráficos de dispersão ou estatisticamente usando testes de homogeneidade de variância, como o teste de Levene.
*   **Independência**: As observações em uma amostra devem ser independentes das observações na outra amostra.

A seguir, vamos realizar algumas verificações utilizando o conjunto de dados birthwt. E vamos mostrar exemplos gráficos para verificar a normalidade e a homogeneidade de variância:

```r
library(ggplot2) # Carregando a biblioteca para gráficos
```

# Explicando o eixo Y (Densidade)

Os valores "0e+00" e "2e-04" no eixo y do gráfico de densidade representam notação científica. Aqui está o significado de cada parte:

**1. "0e+00"**: Isso indica zero elevado à potência de zero, o que é igual a 1. Portanto, "0e+00" é simplesmente 0.

**2. "2e-04"**: Isso indica 2 multiplicado por 10 elevado à potência de -4. Portanto, "2e-04" é igual a 0.0002.

**3. "4e-04"**: Isso indica 4 multiplicado por 10 elevado à potência de -4. Portanto, "4e-04" é igual a 0.0004.

Essa notação é comumente usada para representar números muito pequenos de uma maneira mais compacta. No contexto do gráfico de densidade, esses valores indicam a escala na qual a densidade está sendo representada. "0e+00" indica que a densidade é muito próxima de zero, enquanto "2e-04" indica que a densidade está em torno de 0.0002. Esses valores são úteis para interpretar a escala relativa das densidades no gráfico.

Adicionando Teste de Igualdade de Variâncias e Teste de Normalidade
**Teste de Igualdade de Variâncias**:


Para testar a igualdade das variâncias entre mães fumantes e não fumantes, iremos usar o teste de Levene. Adicionando o seguinte código ao nosso notebook:

Nós identificamos que devido ao uso inadequado do teste de Levene com uma variável explicativa quantitativa na fórmula do R. A variável em questão, smoke, representa se as mães no estudo eram fumantes ou não fumantes, o que é essencialmente uma classificação categórica.

Para corrigir o problema, nós convertemos a variável smoke em um fator, que é o tipo de dado apropriado no R para representar variáveis categóricas. Isso foi feito utilizando a função as.factor(). A conversão assegura que o teste de Levene interprete corretamente smoke como uma variável categórica, permitindo assim a comparação das variâncias do peso ao nascer entre os dois grupos de mães (fumantes e não fumantes).

Após essa correção, nós executamos novamente o teste de Levene, desta vez com a variável smoke tratada corretamente como um fator. Isso permitiu a realização do teste sem erros, possibilitando avaliar se as variâncias dos pesos ao nascer entre bebês de mães fumantes e não fumantes são iguais ou não.

```r
# Convertendo 'smoke' para um fator se ainda não for
birthwt$smoke <- as.factor(birthwt$smoke)

# Realizando o teste de Levene para igualdade de variâncias
leveneTest(bwt ~ smoke, data = birthwt)
```

O teste de Levene revelou que o valor F é aproximadamente 1.456, com um valor-p de cerca de 0.229. Isso indica que não há evidências suficientes para afirmar que as variâncias dos pesos ao nascer dos bebês de mães fumantes e não fumantes são diferentes. Em outras palavras, as variâncias são estatisticamente iguais de acordo com este teste.

**Teste de Normalidade:**
Para verificar a normalidade dos dados em ambos os grupos, iremos usar o teste de Shapiro-Wilk. Adicionando o seguinte código:

```r
# Teste de Shapiro-Wilk para normalidade
shapiro.test(birthwt$bwt[birthwt$smoke == 0])  # Para mães não fumantes
shapiro.test(birthwt$bwt[birthwt$smoke == 1])  # Para mães fumantes
```

**Visualização da Normalidade com QQ Plots:**
Para visualizar a normalidade, usamos gráficos QQ plot. Adicionando este código:

```r
qqnorm(birthwt$bwt[birthwt$smoke == 0])
qqline(birthwt$bwt[birthwt$smoke == 0])
qqnorm(birthwt$bwt[birthwt$smoke == 1])
qqline(birthwt$bwt[birthwt$smoke == 1])
```

Nós utilizamos os gráficos QQ plot (Quantile-Quantile plots) para visualizar e avaliar a normalidade das distribuições dos pesos ao nascer (variável bwt) para dois grupos distintos: mães fumantes e mães não fumantes. Este tipo de gráfico é uma ferramenta poderosa para verificar a aderência de uma distribuição de dados a uma distribuição teórica, neste caso, a distribuição normal.

Aqui está o que aconteceu em cada passo:

*   ***qqnorm(birthwt$bwt[birthwt$smoke == 0]):*** Esse comando gerou um QQ plot para os pesos ao nascer dos bebês de mães não fumantes. O gráfico colocou os quantis (percentis) dos dados observados no eixo Y e os quantis correspondentes de uma distribuição normal no eixo X. Se os dados seguissem uma distribuição normal, os pontos no gráfico se alinhariam aproximadamente em uma linha reta.

*   ***qqline(birthwt$bwt[birthwt$smoke == 0]):*** Esse comando adicionou uma linha de referência ao gráfico QQ anterior. Esta linha representa o local onde os pontos deveriam cair se os dados seguissem perfeitamente uma distribuição normal. A aderência dos pontos a essa linha é um indicativo da normalidade dos dados.

*   **Repetindo para mães fumantes:** Os comandos qqnorm(birthwt$bwt[birthwt$smoke == 1]) e qqline(birthwt$bwt[birthwt$smoke == 1]) realizaram a mesma análise para o grupo de mães fumantes.

Ao avaliar os gráficos QQ plot, nós observamos o quão próximos os pontos estavam da linha. Pontos que se desviavam significativamente da linha indicavam desvios da normalidade. Por exemplo, uma curvatura sistemática nos pontos poderia sugerir distribuições com caudas mais leves ou mais pesadas do que a normal. Essa análise visual, combinada com os resultados do teste de Shapiro-Wilk, forneceu uma compreensão abrangente da normalidade dos dados em cada grupo.
